```python
####UNIVARIATE ANALYSIS###
#######Univariate analysis is the statistical analysis of a single variable.######
jowa['Flight Number'].hist()
plt.title(' Flight Number Distribution')
plt.xlabel('Flight Number')
plt.ylabel('Frequency')
plt.show()
```

The histogram displays the frequency distribution of the 'Flight Number' values, which shows how many times each unique 'Flight Number' value appears in the dataset.

The height of each bar in the histogram corresponds to the frequency of the corresponding 'Flight Number' value.

The x-axis of the histogram represents the 'Flight Number' values, and the y-axis represents the frequency of each value. The height of each bar in the histogram corresponds to the frequency of the corresponding 'Flight Number' value.

```python
flightnumber_data = jowa['Flight Number']
plt.figure(figsize=(8,6))
plt.scatter(flightnumber_data, [0]*len(flightnumber_data))
plt.title('Flight Number Distribution')
plt.xlabel('Flight Number')
plt.ylabel('Constant Value')
plt.show()
```

The scatter plot shows the distribution of the 'Flight Number' values along the x-axis, with a horizontal line of points at y=0. This can help to visualize the distribution of the data and identify any patterns or trends in the data.

```python
####BIVARIATE ANALYSIS It involves the analysis of two variables#####
jowa.head(10)
```

```python
jowa['Time Difference (Minutes)'] = pd.to_numeric(jowa['Time Difference (Minutes)'], errors='coerce')
jowa['Time Difference (Minutes)'].mean()
```

```python
flightnumber_data = 'Flight Number'
timedifference_data = 'Time Difference (Minutes)'

# a scatter plot of the two columns
plt.scatter(jowa[flightnumber_data], jowa[timedifference_data])
# these are our axis labels and a title
plt.xlabel(flightnumber_data)
plt.ylabel(timedifference_data)
plt.title(f'Scatter Plot of {flightnumber_data} vs. {timedifference_data}')
# Show the plot
plt.show()
```

this scatter plot displays the time difference (minutes)versus the flight number. thus showing how many minutes were in between each flight

```python
flightnumber_data = 'Flight Number'
timedifference_data = 'Time Difference (Minutes)'
plt.hist(jowa[flightnumber_data], alpha=0.5, label=flightnumber_data)
plt.hist(jowa[timedifference_data], alpha=0.5, label=timedifference_data)
plt.xlabel(flightnumber_data)
plt.ylabel('timedifference_data')
plt.title(f'Histogram of {flightnumber_data} and {timedifference_data}')
plt.legend()
plt.show()
```

the above histogram visualizes the data for both the flight number and time difference against frequency

```python
jowa['Taxi-in Time (Minutes)'] = pd.to_numeric(jowa['Taxi-in Time (Minutes)'], errors='coerce')
jowa['Taxi-in Time (Minutes)'].mean()
```

```python
flightnumber_data = 'Flight Number'
taxitime_data = 'Taxi-in Time (Minutes)'
plt.plot(jowa[flightnumber_data], jowa[taxitime_data])
plt.xlabel(flightnumber_data)
plt.ylabel(taxitime_data)
plt.title(f'Line Graph of {flightnumber_data} vs. {taxitime_data}')
plt.show()
```

The line graph can help you visualize the relationship between the two variables and identify any patterns or trends in the data.

```python
###MULTIVARIATE ANALYSIS#####
###Multivariate Analysis, involves evaluating multiple variables (more than two) to identify any possible association among them.##
jowa = jowa.select_dtypes(include=[np.number])
sns.heatmap(jowa.corr(), annot=True, cmap='coolwarm')
plt.xlabel('Variables')
plt.ylabel('Variables')
plt.title('Heatmap of Correlation Coefficients')
plt.show()
```
