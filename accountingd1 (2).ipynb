```python
# Univariate Analysis: Histogram of Median Income
joan = joan[1:]

joan['Years Undergraduate Study'] = pd.to_numeric(joan['Years Undergraduate Study'], errors='coerce')

plt.figure(figsize=(10, 6))
sns.histplot(joan['Years Undergraduate Study'].dropna(), bins=20)
plt.title('Distribution of Median Income')
plt.xlabel('Median Income')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

```python
# Bivariate Analysis: Plotting  Scatter plot of Median Age vs. Median Income
plt.figure(figsize=(10, 6))
sns.scatterplot(x=joan['Years Undergraduate Study'].astype(float), y=joan['Years of Service'].astype(float))
plt.title('Relationship Between Years Undergraduate Study Age and Median Income')
# plt.xlabel('Median Age')
# plt.ylabel('Median Income')
plt.grid(True)
plt.show()
```

This scatter plot allows you to visually assess the relationship between the "Years Undergraduate Study" and "Years of Service" variables. Each point on the plot represents an observation in the dataset, and its position is determined by the values of the two variables. Scatter plots are useful for identifying patterns, trends, and potential correlations between two numerical variables.

```python
# Trivariate Analysis: 3D Scatter plot of Median Age, Median Income, and Average Bank Balance
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

# Converting "Average Bank Balance" to numeric, handling any non-numeric values
joan['Years of Service'] = pd.to_numeric(joan['Years of Service'], errors='coerce')

ax.scatter(joan['Gender'].astype(float),
           joan['Years Undergraduate Study'].astype(float),
           joan['Years of Service'].astype(float))

ax.set_title('3D Relationship Among Gender, Years Undergraduate Study, and Years of Service')
ax.set_xlabel('Gender')
ax.set_ylabel('Years Undergraduate Study')
ax.set_zlabel('Years of Service')
plt.show()
```

The scatter plot visualizes the relationship among the specified variables in a three-dimensional space. Each data point is represented by a marker, and its position is determined by the values of the three variables. This type of plot can help identify patterns or trends in the data and understand how the variables are distributed in the 3D space.
